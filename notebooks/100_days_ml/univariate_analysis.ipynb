Import pandas and seaborn, then load the dataset for analysis.

```python
import pandas as pd
import seaborn as sns
df = pd.read_csv("../../datasets/train.csv")
df.head(6)
```

Inspect the dataframe columns to understand available features.

```python
df.columns
```

Compute correlations with the target variable `Survived`.

```python
df.corr(numeric_only=True)["Survived"]
```

Plot the distribution of survival labels.

```python
df["Survived"].value_counts().plot(kind="bar")
```

Plot counts of embarkation ports to compare boarding locations.

```python
sns.countplot(df["Embarked"])
```

Use a box plot to visualize the fare distribution and highlight outliers.

```python
sns.boxplot(x=df["Fare"])
```

Compute age skewness to quantify asymmetry in the distribution.

```python
df["Age"].skew()
```

Plot passenger gender proportions with a pie chart.

```python
df["Survived"].value_counts().plot(kind="pie")
```
