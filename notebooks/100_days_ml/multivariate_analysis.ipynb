Visualize the relationship between total bill amount and tip using a scatter plot

## Load Titanic Dataset
Load the titanic dataset from CSV file and display the first 5 rows

```python
titanic = pd.read_csv("../../datasets/train.csv")
titanic.head(5)
```

## Bar Plot: Average Age by Passenger Class
Analyze the average age of passengers grouped by passenger class

```python
sns.barplot(x=titanic["Pclass"], y=titanic["Age"])
```

## View Dataset Columns
Display all column names in the titanic dataset

```python
titanic.columns
```

## Bar Plot: Age by Passenger Class and Sex
Analyze average age by passenger class with sex differentiation using hue

```python
sns.barplot(x = titanic["Pclass"], y =titanic["Age"], hue= titanic["Sex"])
```

## Bar Plot: Fare by Passenger Class and Sex
Analyze average fare by passenger class with sex differentiation

```python
sns.barplot(x = titanic["Pclass"], y =titanic["Fare"], hue= titanic["Sex"])
```
